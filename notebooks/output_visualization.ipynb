SCARlink creates visualizations for gene-models by plotting scATAC-seq and scRNA-seq grouped by cell type annotations provided by the user. The cell type annotations must be present in the Seurat/ArchR objects used during preprocessing. The plots generated in this notebook can also be generated at the command line using:

```
scarlink_plot -o human_cortex --genes SATB2,SOX6 --celltype celltype

````python
import scarlink.src.visualization as scv
````

````python
dirname = 'human_cortex/'
````

Call `get_scarlink_output` to collect all the genes that the SCARlink was run on. Then call `plot_scarlink_output` to plot the SCARlink visualizations for a set of genes. Here, we group the cells based on `celltype`. The learned regeression coefficients are plotted at the bottom of the scATAC-seq data and indicate the overall importance of each tile. Cell-type-specific scaled Shapley values are plotted as blue dots below each type that highlight the importance of the tile for predicting gene expression in the given cell type.

````python
scarlink_out = scv.get_scarlink_output(dirname)
````

````python
scv.plot_scarlink_output(scarlink_out, genes=['SATB2', 'SOX6'], celltype='celltype')
````
